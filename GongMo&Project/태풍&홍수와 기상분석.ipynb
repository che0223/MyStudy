```python
from datetime import datetime

def findtime(datelist1, datelist2):
# 날짜 입력 받기
    date_str1 = "-".join(map(str, datelist1))
    date_str2 = "-".join(map(str, datelist2))
    return int((datetime.strptime(date_str1, "%Y-%m-%d-%H") - datetime.strptime(date_str2, "%Y-%m-%d-%H")).total_seconds()//3600)

type(datetime.strptime('2013-1-1-1', "%Y-%m-%d-%H") - datetime.strptime('2013-1-1-1', "%Y-%m-%d-%H"))
```
