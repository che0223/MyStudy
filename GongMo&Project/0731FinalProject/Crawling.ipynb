# E-mart 품명, 상품 이미지 수집

```python
driver = webdriver.Chrome() # 웹 브라우저 가동
driver.get('https://emart.ssg.com/') # 해당 html 이동
```

```python
from selenium.webdriver.common.by import By # find_element_by_xpath에서 find_element 함수로 변경
from selenium.webdriver.common.keys import Keys# find_element.click() 이 안될 경우 사용
from selenium.webdriver.common.action_chains import ActionChains # 액션을 묶어서 사용 -> 뭔가 편하다

# driver.find_element(by=By.XPATH, value = r'//*[@id="skip_gnb"]/div/div[1]/div/div/ul/li[10]/a').click() # 오류 발생
# driver.find_element(by=By.XPATH, value = r'//*[@id="skip_gnb"]/div/div[1]/div/div/ul/li[10]/a').send_keys(Keys.ENTER) # 오류발생
# 여기서 가정을 하게된다 : 혹시 마우스 드래그를 할때 새로운 html 문구가 생기는 형식이 아닐까? 그래서 새로운 html 문구가 안생긴 상태로 클릭을 해서 오류가 생기지 않았을까?

pead = driver.find_element(by=By.XPATH, value= r'//*[@id="skip_gnb"]/div/div[1]/button')
button = driver.find_element(by=By.XPATH, value = r'//*[@id="skip_gnb"]/div/div[1]/div/div/ul/li[10]/a')

actions = ActionChains(driver)
actions.move_to_element(pead).perform() # 해당 위치로 드래그
actions.click(on_element=button).perform() # 클릭
# 훌륭하게 성공했다.
```

```python
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException

product_list = []
price_list = []

for i in range(1, 81):
    try:
        name = driver.find_element(by=By.XPATH, value = r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[1]/span[2]').text
    except NoSuchElementException as e: # 브랜드 명 없이 상품명만 있을 경우
        name = driver.find_element(by=By.XPATH, value = r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[1]/span').text

    try:
        price = driver.find_element(by=By.XPATH, value = r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[2]/div[2]/div[2]/em').text
    except NoSuchElementException as e: # 브랜드 명 없이 상품명만 있을 경우
        price = driver.find_element(by=By.XPATH, value = r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[2]/div/div[1]/em').text
    
    product_list.append(name)
    price_list.append(price)
    print(name, price)
```

```python
import pandas as pd

df = pd.DataFrame({'상품명':product_list, '가격':price_list})
```

```python
df.to_csv('./result.csv', index=False)
```

```python
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException, TimeoutException
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

product_list = []
price_list = []

for j in range(1, 10):
    driver.find_element(by=By.XPATH, value = f'//*[@id="area_itemlist"]/div[2]/div/a[{j}]').click()
    # time.sleep(1) # 페이지가 나타나기 전에 읽는 경우 에러가 뜨므로 서버에서 데이터를 기다리는 처리가 필요
    # driver.implicitly_wait(10) # 페이지가 10초안에 나타나면 바로 다음페이지
    try:
        WebDriverWait(driver, 8).until(EC.presence_of_element_located((By.XPATH, r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[1]/span[2]')))
    except TimeoutException:
        WebDriverWait(driver, 8).until(EC.presence_of_element_located((By.XPATH, r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[1]/span')))
    # time.sleep(1) 보다 좀더 개선된 방법 : 그 페이지가 실제 로드 될때까지 기다림(기다리는 최대 시간도 정할 수 있음)
    # implicitly_wait로도 페이지 로딩되기전 에러 발생가능성도 있음 => WebDriverWait 이용
    
    for i in range(1, 81):
        try:
            name = driver.find_element(by=By.XPATH, value = r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[1]/span[2]').text
        except NoSuchElementException as e: # 브랜드 명 없이 상품명만 있을 경우
            name = driver.find_element(by=By.XPATH, value = r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[1]/span').text
        try:
            price = driver.find_element(by=By.XPATH, value = r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[2]/div[2]/div[2]/em').text
        except NoSuchElementException as e: # 브랜드 명 없이 상품명만 있을 경우
            price = driver.find_element(by=By.XPATH, value = r'//*[@id="ty_thmb_view"]/ul/li['+str(i)+']/div/div/div[2]/div[1]/a/div[2]/div/div[1]/em').text

        product_list.append(name)
        price_list.append(price)
        print(name, price)
```

```python
import pandas as pd
import re

df = pd.DataFrame({'상품명':product_list, '가격':price_list})
df['가격'].apply(re.sub('\d+,\d{3}'))
```
