```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# 0. 브라우저 가동
browser = webdriver.Chrome()
browser.maximize_window() # 창 최대화
```

```python
# 1. 페이지 이동
url = 'https://finance.naver.com/sise/sise_market_sum.naver?&page='
browser.get(url)
```

```python
# 2. 조회 항목 초기화(체크되어 있는 항목 체크 해제)
checkboxes = browser.find_elements(by=By.NAME, value='fieldIds') #'//*[@id="option1"]'
for checkbox in checkboxes:
    if checkbox.is_selected(): # 체크된 상태라면?
        checkbox.click() # 클릭해서 체크해제

# 3. 조회 항목 체크(원하는 항목)
items_to_select = ['영업이익', '자산총계', '매출액']
for checkbox in checkboxes:
    parent = checkbox.find_element(by=By.XPATH, value='..') # 부모 element
    label = parent.find_element(by=By.TAG_NAME, value = 'label')
    
    if label.text in items_to_select:
        checkbox.click()
# 4. 적용하기 버튼 클릭
# <a href="javascript:fieldSubmit()"><img src="https://ssl.pstatic.net/imgstock/images5/btn_apply.gif" alt="적용하기" width="55" height="18"></a>
# btn_apply = browser.find_element(By.XPATH, value = '//*[@id="contentarea_left"]/div[2]/form/div/div/div/a[1]')
btn_apply = browser.find_element(By.XPATH, value = '//a[@href="javascript:fieldSubmit()"]') # //는 전체 html태그에서 찾는다는 뜻, a는 a태그, @속성 = 값
btn_apply.click()
```

```python
# 5. 데이터 추출
df = pd.read_html(browser.page_source)[1]
df = df.dropna(axis=0, how='all')
df['N'] = df['N'].astype(int)
df = df.set_index('N')

df = df.dropna(axis=1, how='all')
df
```

```python
import os

# 6. 파일 저장
f_name = 'sise.csv'
if os.path.exists(f_name): # 파일이 있다면? 헤더 제외
    df.to_csv(f_name, encoding='utf-8-sig', index=False, mode='a', header=False) # append
else:
    df.to_csv(f_name, encoding='utf-8-sig', index=False)

print(f'{idx} 페이지 완료')
```
